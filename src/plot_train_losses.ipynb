```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_csv = "2022-02-03_22-21-11/train_metrics.csv"
```

```python
df_train = pd.read_csv(file_csv)
```

```python
df_train.head()
```

```python
fig = plt.figure(figsize=(12, 9))
plt.plot(np.arange(1, 1+len(df_train.loss_gen_l1)), df_train.loss_gen_l1, "o-")
plt.grid()
plt.xlabel("epoch")
plt.ylabel("Avg. L1 loss")
plt.title("Generator's avg. L1 loss vs epochs")
fig.show()
```

```python
fig = plt.figure(figsize=(12, 9))
plt.plot(np.arange(1, 1+len(df_train.loss_gen_gan)), df_train.loss_gen_gan, "o-", label="Generator's BCE loss to fool Discriminator")
plt.plot(np.arange(1, 1+len(df_train.loss_dis_real)), df_train.loss_dis_real, "o-", label="Discriminator's BCE loss for identifying real")
plt.plot(np.arange(1, 1+len(df_train.loss_dis_fake)), df_train.loss_dis_fake, "o-", label="Discriminator's BCE loss for identifying fake")
plt.grid()
plt.xlabel("epoch")
plt.ylabel("Avg. BCE loss")
plt.legend()
plt.title("Comparison of Generator's and Discriminator's avg. BCE losses vs epochs")
fig.show()
```
